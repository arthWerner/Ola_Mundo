```python
salario_mensal = float(input("Digite o seu salário mensal: R$ "))
horas_mes = float(input("Digite a quantidade de horas trabalhadas no mês: "))

valor_hora = salario_mensal / horas_mes

print(f"O seu valor por hora é: R$ {valor_hora:.2f}")
```
